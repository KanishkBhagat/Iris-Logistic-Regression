```python
#importing necessary libraries

import pandas as pd                                                   # Data Manipulation
import numpy as np                                                    # Array Manipulation
import seaborn as sns                                                 # Data Visualization
import matplotlib.pyplot as plt                                       # Data Visualization
import warnings                                                       # To ignore warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler                      # For normalizing Data
from sklearn.model_selection import train_test_split                  # For Splitting Data
from sklearn.linear_model import LogisticRegression                   # For importing our model
from sklearn.metrics import confusion_matrix,classification_report    # model accuracy
```

```python
# importing our dataset

df=pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
# Checking null Values

df.isnull().sum()
```

```python
# Creating Feature and Target Variable

x=df.drop(['Species'],axis=1)
y=df['Species']
```

```python
# Splitting String to obtain species names

y=y.str.split('-',expand=True)[1]
```

```python
# Converting Species Names into numeric values

label={'setosa':0,'versicolor':1,'virginica':2}
y=y.map(label)
```

```python
# Checking the correlation between variables

sns.heatmap(x.corr(),annot=True)
```

```python
# Visualizing the Distribution of Variables

plt.scatter(x['SepalLengthCm'],x['PetalLengthCm'],c='g')
plt.scatter(x['SepalWidthCm'],x['PetalWidthCm'])
plt.show()
```

```python
for i in x.columns[1:]:
    sns.displot(x[i],kind='kde')
    plt.title(i)
```

```python
sns.pairplot(df,hue='Species')
```

```python
# Splitting Data into Training and Testing split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=30)
```

```python
# Standardizing x train, and x test data

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Creating our model

reg=LogisticRegression()
model=reg.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# Checking model accuracy

print(confusion_matrix(y_test,y_pred))
print('-'*90)
print(classification_report(y_test,y_pred))
```
